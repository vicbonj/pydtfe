```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Start by importing the module pydtfe

```python
from pydtfe.pydtfe import Pydtfe
```

## Then create a fake distribution of points

```python
x, y = np.random.random((2, 100000)) * 20 - 10
xx, yy = np.random.randn(2, 10000)
xxx = np.hstack([x, xx])
yyy = np.hstack([y, yy])
```

## One can look at the distribution you've just made

```python
plt.figure(figsize=(8,8))
plt.plot(xxx, yyy, '+', markersize=0.5)
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
```

## Define a size for your grid...

```python
size=1000
```

## ... and run the function map_dtfe

```python
p = Pydtfe(xxx, yyy, xmin=-10, xmax=10, ymin=-10, ymax=10, xsize=size, ysize=size)
```

## You can plot your grid with points overlapping

```python
p.show(log_scale=True, show_points=False)
```
